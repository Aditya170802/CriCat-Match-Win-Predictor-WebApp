# normalization of each player and scoring them:

```python
masterdatat20 = pd.read_excel('MasterT20.xlsx')
matches_std = masterdatat20['Matches'].std()
matches_mean = masterdatat20['Matches'].mean()
runs_std = masterdatat20['Runs'].std()
runs_mean = masterdatat20['Runs'].mean()
bt_avg_std = masterdatat20['Bat Avg'].std()
bt_avg_mean = masterdatat20['Bat Avg'].mean()
Wkts_std = masterdatat20['Wkts'].std()
Wkts_mean = masterdatat20['Wkts'].mean()
blavg_std = masterdatat20['Bowl Avg'].std()
blavg_mean = masterdatat20['Bowl Avg'].mean()
```

masterdatat20['Normal Score'] = 0

```python
masterdatat20 = masterdatat20.set_index("Name", drop = False)
```

```python
masterdatat20.fillna(0, inplace=True)
```

```python
def normcode(na):
    score = 0.0
    for i in ['Matches', 'Runs', 'Bat Avg', 'Wkts', 'Bowl Avg']:
        score += (masterdatat20.loc[na, i] - masterdatat20[i].mean())/masterdatat20[i].std()
    return score

```

```python
Normal_Scores=[]
```

```python
for i in masterdatat20['Name']:
        Normal_Scores.append(round(normcode(i), 3))
```

```python
masterdatat20['Normal_Scores'] = Normal_Scores
```
